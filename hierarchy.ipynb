```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
data=pd.read_csv("D:\\DATA_science\\Data_sets\\Universities.csv")
```

```python
data.head()
```

```python
def norm_fun(i):
    x=(i-i.min())/(i.max()-i.min())
    return(x)
```

```python
norm_data=norm_fun(data.iloc[:,1:])
```

```python
norm_data.head()
```

```python
dendrogram=sch.dendrogram(sch.linkage(norm_data,method='complete'))
```

```python
hc=AgglomerativeClustering(n_clusters=4,affinity='euclidean',linkage='complete')
```

```python
y_hc=hc.fit_predict(norm_data)
#Clusters=pd.DataFrame(y_hc,columns=['Clusters'])
#Clusters
```

```python
norm_data['Clusterid']=hc.labels_
```
