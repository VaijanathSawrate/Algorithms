```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans
```

```python
data=pd.read_csv("D:\\DATA_science\\Data_sets\\Universities.csv")
```

```python
from sklearn.preprocessing import StandardScaler 
scaler=StandardScaler()
scaled_data_df=scaler.fit_transform(data.iloc[:,1:])
```

```python
wcss=[]
for i in range (1,11):
    kmeans=KMeans(n_clusters=i,random_state=0)
    kmeans.fit(scaled_data_df)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1,11),wcss)
plt.title('Elbow Method')
plt.xlabel('Number of Cludters')
plt.ylabel('wcss')
plt.show()
```

```python
from sklearn.cluster import KMeans
cluster_new=KMeans(4,random_state=42)
cluster_new.fit(scaled_data_df)
```

```python
cluster_new.labels_
```

```python
data['clusterid_new']=cluster_new.labels_
```

```python
cluster_new.cluster_centers_
```
